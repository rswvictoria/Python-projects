```python
import numpy as np
import pandas as pd
ab = pd.read_csv('data_v5.csv',encoding='windows-1252')
ab.head()
```

```python
ab_time = pd.get_dummies(ab['New_time'],prefix='time')
ab_time.head()

ab = pd.concat([ab, ab_time], axis=1)
ab.head()
```

```python
dummydata = ab.ix[:,10:]
dummydata.head()


```

```python
X_labels = [c for c in dummydata.columns if c not in ['Severity2']]
X = dummydata.ix[:,X_labels]
Y = dummydata['Severity2']
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(X,Y)
```

```python
estimated_coefficients = pd.DataFrame(columns=['features','estimated coefficients'])
estimated_coefficients['features']=X.columns
estimated_coefficients['estimated coefficients']=lm.coef_
estimated_coefficients
```

```python
#scores
lm.score(X,Y)
```
